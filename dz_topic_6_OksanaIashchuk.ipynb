## 1. **Import the required libraries.**

```python
# Reproducibility + essentials
import os, random, time, copy
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms, models

from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report

def set_seed(seed=42):
    random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
```

## 2. **Data preparation:**

–ó–∞–≤–∞–Ω—Ç–∞–∂—Ç–µ –Ω–∞–±—ñ—Ä –¥–∞–Ω–∏—Ö Intel Image Classification –∑ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∏ Kaggle.

–†–æ–∑–¥—ñ–ª—ñ—Ç—å –¥–∞–Ω—ñ –Ω–∞ —Ç—Ä–µ–Ω—É–≤–∞–ª—å–Ω–∏–π, –≤–∞–ª—ñ–¥–∞—Ü—ñ–π–Ω–∏–π —Ç–∞ —Ç–µ—Å—Ç–æ–≤–∏–π –Ω–∞–±–æ—Ä–∏. –î–∞–Ω—ñ –≤–∂–µ —Ä–æ–∑–¥—ñ–ª–µ–Ω—ñ –Ω–∞ –Ω–∞–≤—á–∞–ª—å–Ω—ñ —Ç–∞ —Ç–µ—Å—Ç–æ–≤—ñ. –í–∞–º —Å–ª—ñ–¥ –≤—ñ–¥–¥—ñ–ª–∏—Ç–∏ —á–∞—Å—Ç–∏–Ω—É –Ω–∞–≤—á–∞–ª—å–Ω–∏—Ö –¥–∞–Ω–∏—Ö –¥–ª—è –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –≤–∞–ª—ñ–¥–∞—Ü—ñ–π–Ω–æ–≥–æ –Ω–∞–±–æ—Ä—É –¥–∞–Ω–∏—Ö.

–í–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ torchvision.datasets —ñ torch.utils.data.DataLoader –¥–ª—è –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è —Ç–∞ –ø—ñ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–∏—Ö.

–ü–æ–ø–µ—Ä–µ–¥–Ω—è –æ–±—Ä–æ–±–∫–∞ –∑–æ–±—Ä–∞–∂–µ–Ω—å: –ó–∞—Å—Ç–æ—Å—É–π—Ç–µ –Ω–µ–æ–±—Ö—ñ–¥–Ω—ñ –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –∑–º—ñ–Ω–∞ —Ä–æ–∑–º—ñ—Ä—É, –Ω–æ—Ä–º–∞–ª—ñ–∑–∞—Ü—ñ—è) –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é torchvision.transforms.

---
–û–ø–∏—Å:
–î–∞–Ω—ñ –ª–µ–∂–∞—Ç—å —É –ø–∞–ø—Ü—ñ IntelImageClassificationDataset.

–£—Å–µ—Ä–µ–¥–∏–Ω—ñ:

seg_train/seg_train ‚Äî —Ç—Ä–µ–Ω—É–≤–∞–ª—å–Ω—ñ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è,

seg_test/seg_test ‚Äî —Ç–µ—Å—Ç–æ–≤—ñ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è.

–î—ñ–ª–∏–º–æ train –Ω–∞ train (80%) —ñ val (20%).

–í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ –±–∞–∑–æ–≤—ñ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü—ñ—ó: Resize ‚Üí ToTensor ‚Üí Normalize.

```python
# –¢—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü—ñ—ó (Resize -> ToTensor -> Normalize)
train_tfms = transforms.Compose([
    transforms.Resize((150, 150)),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])
test_tfms = transforms.Compose([
    transforms.Resize((150, 150)),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])

# –®–ª—è—Ö–∏ –¥–æ –¥–∞–Ω–∏—Ö
train_dir = "./IntelImageClassificationDataset/seg_train/seg_train"
test_dir  = "./IntelImageClassificationDataset/seg_test/seg_test"

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è train —ñ test
full_train = datasets.ImageFolder(root=train_dir, transform=train_tfms)
test_ds    = datasets.ImageFolder(root=test_dir,  transform=test_tfms)

# –†–æ–∑–¥—ñ–ª–µ–Ω–Ω—è train -> train + val
val_size   = int(0.2 * len(full_train))
train_size = len(full_train) - val_size
train_ds, val_ds = random_split(full_train, [train_size, val_size])

# DataLoader
train_dl = DataLoader(train_ds, batch_size=32, shuffle=True)
val_dl   = DataLoader(val_ds, batch_size=32, shuffle=False)
test_dl  = DataLoader(test_ds, batch_size=32, shuffle=False)

# –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞
print("Train samples:", len(train_ds))
print("Val samples:", len(val_ds))
print("Test samples:", len(test_ds))
print("Classes:", full_train.classes)
```

## 3. **–°—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ: –í–∞—Ä—ñ–∞–Ω—Ç A: –ü—Ä–æ—Å—Ç–∞ –∑–≥–æ—Ä—Ç–∫–æ–≤–∞ –Ω–µ–π—Ä–æ–Ω–Ω–∞ –º–µ—Ä–µ–∂–∞.**

–°—Ç–≤–æ—Ä—ñ—Ç—å –≤–ª–∞—Å–Ω—É –∞—Ä—Ö—ñ—Ç–µ–∫—Ç—É—Ä—É CNN –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é torch.nn.

–í–∏–∑–Ω–∞—á—Ç–µ —à–∞—Ä–∏ –∑–≥–æ—Ä—Ç–∫–∏ (Conv2d), –∞–∫—Ç–∏–≤–∞—Ü—ñ—ó (ReLU), –ø—É–ª—ñ–Ω–≥—É (MaxPool2d) —Ç–∞ –ø–æ–≤–Ω–æ–∑–≤'—è–∑–Ω—ñ —à–∞—Ä–∏ (Linear).

–û–ø–∏—Å:

–ó–∞–≤–¥–∞–Ω–Ω—è ‚Äî –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—è –∑–æ–±—Ä–∞–∂–µ–Ω—å –Ω–∞ 6 –∫–ª–∞—Å—ñ–≤ (–∫–∞—Ç–µ–≥–æ—Ä—ñ–π –ø—Ä–∏—Ä–æ–¥–Ω–∏—Ö —Å—Ü–µ–Ω).

–ë—É–¥—É—î–º–æ –≤–ª–∞—Å–Ω—É –ø—Ä–æ—Å—Ç—É –∑–≥–æ—Ä—Ç–∫–æ–≤—É –Ω–µ–π—Ä–æ–Ω–Ω—É –º–µ—Ä–µ–∂—É (CNN) ¬´–∑ –Ω—É–ª—è¬ª.

–ê—Ä—Ö—ñ—Ç–µ–∫—Ç—É—Ä–∞:

–ö—ñ–ª—å–∫–∞ –±–ª–æ–∫—ñ–≤ Conv2d ‚Üí ReLU ‚Üí MaxPool2d –¥–ª—è –≤–∏–¥—ñ–ª–µ–Ω–Ω—è –æ–∑–Ω–∞–∫.

–ù–∞–ø—Ä–∏–∫—ñ–Ω—Ü—ñ ‚Äî Linear —à–∞—Ä–∏ –¥–ª—è –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—ó.

–í–∏—Ö—ñ–¥–Ω–∏–π —à–∞—Ä –º–∞—î —Ä–æ–∑–º—ñ—Ä = –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∫–ª–∞—Å—ñ–≤ (len(full_train.classes)).

```python
class SimpleCNN(nn.Module):
    def __init__(self, num_classes):
        super(SimpleCNN, self).__init__()
        
        self.conv_layers = nn.Sequential(
            # Block 1
            nn.Conv2d(3, 32, kernel_size=3, padding=1),  # 3 –∫–∞–Ω–∞–ª–∏ -> 32 —Ñ—ñ–ª—å—Ç—Ä–∏ (—Å—Ç–∞–Ω–¥–∞—Ä—Ç –¥–ª—è –º–∞–ª–µ–Ω—å–∫–∏—Ö CNN). –Ω–∞ —Ä–µ–∞–ª—å–Ω–∏—Ö –ø—Ä–æ–µ–∫—Ç–∞—Ö –¥–æ—Ü—ñ–ª—å–Ω–æ –∑–±—ñ–ª—å—à–∏—Ç–∏
            nn.ReLU(),
            nn.MaxPool2d(2),  # –∑–º–µ–Ω—à–µ–Ω–Ω—è —Ä–æ–∑–º—ñ—Ä—É –≤–¥–≤—ñ—á—ñ
            
            # Block 2
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            # Block 3
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        
        self.fc_layers = nn.Sequential(
            nn.Flatten() # —Ü–µ —Å—Ç–∞—î –≤–µ–∫—Ç–æ—Ä–æ–º –¥–æ–≤–∂–∏–Ω–∏ 128 √ó 18 √ó 18 = 41 472 . —Ä–æ–∑–º—ñ—Ä –∑–∞–ª–µ–∂–∏—Ç—å –≤—ñ–¥ –≤—Ö—ñ–¥–Ω–æ–≥–æ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è (150x150 -> 18x18 –ø—ñ—Å–ª—è 3 –ø—É–ª—ñ–Ω–≥—ñ–≤)
            nn.Linear(128 * 18 * 18, 256),  # —Å–ø–æ—á–∞—Ç–∫—É —Å—Ç–∏—Å–∫–∞—î–º–æ –¥–æ 256 –Ω–µ–π—Ä–æ–Ω—ñ–≤. 
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes) # –í–∏—Ö—ñ–¥–Ω–∏–π —à–∞—Ä ‚Üí 6 –∫–ª–∞—Å—ñ–≤ (–±–æ —É –Ω–∞—Å 6 –∫–∞—Ç–µ–≥–æ—Ä—ñ–π: buildings, forest, glacier, mountain, sea, street)
        )
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fc_layers(x)
        return x

# –Ü–Ω—ñ—Ü—ñ–∞–ª—ñ–∑–∞—Ü—ñ—è –º–æ–¥–µ–ª—ñ
num_classes = len(full_train.classes)
model = SimpleCNN(num_classes).to(device)

print(model)
```

## 4. **–í–∞—Ä—ñ–∞–Ω—Ç B: Transfer Learning**

–Ü–º–ø–æ—Ä—Ç—É–π—Ç–µ –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ –Ω–∞–≤—á–µ–Ω—É –º–æ–¥–µ–ª—å (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, ResNet18) –∑ torchvision.models.

–ú–æ–¥–∏—Ñ—ñ–∫—É–π—Ç–µ –æ—Å—Ç–∞–Ω–Ω—ñ–π —à–∞—Ä –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –∫–ª–∞—Å—ñ–≤ —É –≤–∞—à–æ–º—É –∑–∞–≤–¥–∞–Ω–Ω—ñ.

–û–ø–∏—Å:

–ë–µ—Ä–µ–º–æ ResNet18 –∑ torchvision.models –∑ –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ –Ω–∞–≤—á–µ–Ω–∏–º–∏ –≤–∞–≥–∞–º–∏.

–∑–∞–º–æ—Ä–æ–∂—É—î–º–æ backbone, —â–æ–± –≤—á–∏—Ç–∏—Å—å –ª–∏—à–µ –æ—Å—Ç–∞–Ω–Ω—ñ–º —à–∞—Ä–æ–º.

–ó–∞–º—ñ–Ω—é—î–º–æ —Ñ—ñ–Ω–∞–ª—å–Ω–∏–π fc –Ω–∞ Linear(..., num_classes).

```python

# num_classes = len(full_train.classes)  # 6 –¥–ª—è Intel dataset

def build_resnet18(num_classes: int, freeze_backbone: bool = True):
    # –∑–∞–≤–∞–Ω—Ç–∞–∂—É—î–º–æ –º–æ–¥–µ–ª—å –∑ –ø—Ä–µ–¥—Ç—Ä–µ–Ω–æ–º
    model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
    
    # –∑–∞–º–æ—Ä–æ–∂—É—î–º–æ –≤—Å—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏ –±–µ–∫–±–æ–Ω—É
    if freeze_backbone:
        for p in model.parameters():
            p.requires_grad = False
    
    # –∑–∞–º—ñ–Ω—é—î–º–æ –æ—Å—Ç–∞–Ω–Ω—ñ–π –ø–æ–≤–Ω–æ–∑–≤'—è–∑–Ω–∏–π —à–∞—Ä –ø—ñ–¥ –Ω–∞—à—ñ –∫–ª–∞—Å–∏
    in_features = model.fc.in_features
    model.fc = nn.Linear(in_features, num_classes)
    return model

# —Å—Ç–≤–æ—Ä—é—î–º–æ –º–æ–¥–µ–ª—å —ñ –ø–µ—Ä–µ–Ω–æ—Å–∏–º–æ –Ω–∞ –¥–µ–≤–∞–π—Å
model = build_resnet18(num_classes=num_classes, freeze_backbone=True).to(device) # freeze_backbone=True ‚Äî —à–≤–∏–¥–∫–∏–π —Å—Ç–∞—Ä—Ç (–≤—á–∏–º–æ –ª–∏—à–µ –Ω–æ–≤–∏–π fc), –∑–∞–∑–≤–∏—á–∞–π —É–∂–µ –¥–∞—î –≤–∏—Å–æ–∫—É —Ç–æ—á–Ω—ñ—Å—Ç—å.


learnable_params = [p for p in model.parameters() if p.requires_grad]
print("Model ready (ResNet18 TL). Trainable params:",
      sum(p.numel() for p in learnable_params))
```

## 5. **–í–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó –≤—Ç—Ä–∞—Ç —Ç–∞ –æ–ø—Ç–∏–º—ñ–∑–∞—Ç–æ—Ä–∞:**

–í–∏–±–µ—Ä—ñ—Ç—å –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω—É —Ñ—É–Ω–∫—Ü—ñ—é –≤—Ç—Ä–∞—Ç (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, CrossEntropyLoss). –û–±“ë—Ä—É–Ω—Ç—É–π—Ç–µ –≤–∏–±—ñ—Ä —Ñ—É–Ω–∫—Ü—ñ—ó –≤—Ç—Ä–∞—Ç.

–í–∏–±–µ—Ä—ñ—Ç—å –æ–ø—Ç–∏–º—ñ–∑–∞—Ç–æ—Ä (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, Adam –∞–±–æ SGD).

–û–ø–∏—Å:

–û—Å–∫—ñ–ª—å–∫–∏ —Ü–µ –±–∞–≥–∞—Ç–æ–∫–ª–∞—Å–æ–≤–∞ –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—è, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∏–º –≤–∏–±–æ—Ä–æ–º —î CrossEntropyLoss. 
–í–æ–Ω–∞ –ø–æ—î–¥–Ω—É—î LogSoftmax + NLLLoss, —Ç–æ–º—É –Ω–µ –ø–æ—Ç—Ä—ñ–±–Ω–æ —è–≤–Ω–æ –¥–æ–¥–∞–≤–∞—Ç–∏ Softmax —É –º–æ–¥–µ–ª—å.

–î–ª—è –æ–ø—Ç–∏–º—ñ–∑–∞—Ü—ñ—ó –º–æ–∂–Ω–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ Adam, –±–æ –≤—ñ–Ω –¥–æ–±—Ä–µ –ø—Ä–∞—Ü—é—î –π —à–≤–∏–¥—à–µ —Å—Ö–æ–¥–∏—Ç—å—Å—è, –Ω—ñ–∂ –∑–≤–∏—á–∞–π–Ω–∏–π SGD.

–ì—ñ–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∏:

lr = 1e-3 ‚Äî –∫–ª–∞—Å–∏—á–Ω–µ —Å—Ç–∞—Ä—Ç–æ–≤–µ –∑–Ω–∞—á–µ–Ω–Ω—è –¥–ª—è Adam,

weight_decay = 1e-4 ‚Äî –¥–æ–¥–∞—î –ª–µ–≥–∫—É L2-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü—ñ—é, —â–æ–± —É–Ω–∏–∫–∞—Ç–∏ –ø–µ—Ä–µ–Ω–∞–≤—á–∞–Ω–Ω—è.

```python
criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)

print("Loss function:", criterion)
print("Optimizer:", optimizer)
```

## 6. **–ù–∞–≤—á–∞–Ω–Ω—è –º–æ–¥–µ–ª—ñ:**

–ù–∞–ø–∏—à—ñ—Ç—å —Ü–∏–∫–ª –Ω–∞–≤—á–∞–Ω–Ω—è –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º PyTorch.

–û–±—á–∏—Å–ª—é–π—Ç–µ –≤—Ç—Ä–∞—Ç–∏, –≤–∏–∫–æ–Ω—É–π—Ç–µ –∑–≤–æ—Ä–æ—Ç–Ω–µ –ø–æ—à–∏—Ä–µ–Ω–Ω—è —Ç–∞ –æ–Ω–æ–≤–ª—é–π—Ç–µ –≤–∞–≥–∏.

–ó–±–µ—Ä—ñ–≥–∞–π—Ç–µ –º–æ–¥–µ–ª—å –∑ –Ω–∞–π–∫—Ä–∞—â–∏–º–∏ –ø–æ–∫–∞–∑–Ω–∏–∫–∞–º–∏ –Ω–∞ –≤–∞–ª—ñ–¥–∞—Ü—ñ–π–Ω–æ–º—É –Ω–∞–±–æ—Ä—ñ.

–û–ø–∏—Å:

–ü—Ä–æ—Ö–æ–¥–∏–º–æ –µ–ø–æ—Ö–∞–º–∏: –¥–ª—è –∫–æ–∂–Ω–æ—ó ‚Äî —Ç—Ä–µ–Ω—É–≤–∞–Ω–Ω—è –Ω–∞ train_dl, –ø–æ—Ç—ñ–º –æ—Ü—ñ–Ω–∫–∞ –Ω–∞ val_dl.

–†–∞—Ö—É—î–º–æ loss —Ç–∞ accuracy –¥–ª—è train/val.

–ó–±–µ—Ä—ñ–≥–∞—î–º–æ –Ω–∞–π–∫—Ä–∞—â—ñ –≤–∞–≥–∏ –∑–∞ –≤–∞–ª—ñ–¥. accuracy.

–í–µ–¥–µ–º–æ —ñ—Å—Ç–æ—Ä—ñ—é –¥–ª—è –∫—Ä–∏–≤–∏—Ö –Ω–∞–≤—á–∞–Ω–Ω—è.

```python
# Training loop with validation + save best by val accuracy

EPOCHS = 25  

history = {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": []}
best_acc = 0.0
best_state = None

for epoch in range(1, EPOCHS + 1):
    # ---- TRAIN ----
    model.train()
    train_loss_sum, train_correct, train_total = 0.0, 0, 0

    for xb, yb in train_dl:
        xb, yb = xb.to(device), yb.to(device)

        logits = model(xb)
        loss = criterion(logits, yb)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        train_loss_sum += loss.item() * xb.size(0)
        preds = logits.argmax(1)
        train_correct += (preds == yb).sum().item()
        train_total += xb.size(0)

    train_loss = train_loss_sum / train_total
    train_acc  = train_correct / train_total

    # ---- VALIDATION ----
    model.eval()
    val_loss_sum, val_correct, val_total = 0.0, 0, 0

    with torch.no_grad():
        for xb, yb in val_dl:
            xb, yb = xb.to(device), yb.to(device)

            logits = model(xb)
            loss = criterion(logits, yb)

            val_loss_sum += loss.item() * xb.size(0)
            preds = logits.argmax(1)
            val_correct += (preds == yb).sum().item()
            val_total += xb.size(0)

    val_loss = val_loss_sum / val_total
    val_acc  = val_correct / val_total

    # ---- history + best ----
    history["train_loss"].append(train_loss)
    history["val_loss"].append(val_loss)
    history["train_acc"].append(train_acc)
    history["val_acc"].append(val_acc)

    if val_acc > best_acc:
        best_acc = val_acc
        best_state = {k: v.detach().cpu().clone() for k, v in model.state_dict().items()}

    print(f"Epoch {epoch:02d} | "
          f"train loss {train_loss:.4f} acc {train_acc:.3f} | "
          f"val loss {val_loss:.4f} acc {val_acc:.3f}")

# ---- load & save best ----
if best_state is not None:
    model.load_state_dict({k: v.to(device) for k, v in best_state.items()})
    torch.save(model.state_dict(), "best_intel_classifier.pth")
    print(f"Best val acc: {best_acc:.3f} | Weights saved -> best_intel_classifier.pth")
else:
    print("Warning: best_state is None (–ø–µ—Ä–µ–≤—ñ—Ä, —á–∏ —î –¥–∞–Ω—ñ —É val_dl).")
```

## 7 **–û—Ü—ñ–Ω–∫–∞ –º–æ–¥–µ–ª—ñ:**

–û—Ü—ñ–Ω—ñ—Ç—å –ø—Ä–æ–¥—É–∫—Ç–∏–≤–Ω—ñ—Å—Ç—å –º–æ–¥–µ–ª—ñ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–º—É –Ω–∞–±–æ—Ä—ñ –¥–∞–Ω–∏—Ö.
–û–±—á–∏—Å–ª—ñ—Ç—å –º–µ—Ç—Ä–∏–∫–∏, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥ F1-score —Ç–æ—â–æ. –û–±“ë—Ä—É–Ω—Ç—É–π—Ç–µ –≤–∏–±—ñ—Ä –º–µ—Ç—Ä–∏–∫–∏.


–û–ø–∏—Å:

–ü—ñ—Å–ª—è —Ç—Ä–µ–Ω—É–≤–∞–Ω–Ω—è –π –≤–∏–±–æ—Ä—É –Ω–∞–π–∫—Ä–∞—â–∏—Ö –≤–∞–≥ –º–∏ –ø–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –º–æ–¥–µ–ª—å –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–º—É –Ω–∞–±–æ—Ä—ñ.

–í–∞–∂–ª–∏–≤—ñ –º–µ—Ç—Ä–∏–∫–∏:

Accuracy ‚Äî —á–∞—Å—Ç–∫–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ –∫–ª–∞—Å–∏—Ñ—ñ–∫–æ–≤–∞–Ω–∏—Ö –∑–æ–±—Ä–∞–∂–µ–Ω—å.

Macro F1-score ‚Äî –∑–±–∞–ª–∞–Ω—Å–æ–≤–∞–Ω–∞ –º–µ—Ç—Ä–∏–∫–∞, –≤—Ä–∞—Ö–æ–≤—É—î precision —ñ recall –ø–æ –∫–æ–∂–Ω–æ–º—É –∫–ª–∞—Å—É, –≤–∞–∂–ª–∏–≤–∞ –ø—Ä–∏ –¥–∏—Å–±–∞–ª–∞–Ω—Å—ñ.

Confusion Matrix ‚Äî –ø–æ–∫–∞–∑—É—î, —è–∫—ñ –∫–ª–∞—Å–∏ –º–æ–¥–µ–ª—å –ø–ª—É—Ç–∞—î –º—ñ–∂ —Å–æ–±–æ—é.

```python
# Test evaluation (uses best weights from step 6)

model.eval()
all_preds, all_targets = [], []

with torch.no_grad():
    for xb, yb in test_dl:
        xb = xb.to(device); yb = yb.to(device)
        logits = model(xb)
        preds = logits.argmax(1)
        all_preds.append(preds.cpu().numpy())
        all_targets.append(yb.cpu().numpy())

y_pred = np.concatenate(all_preds)
y_true = np.concatenate(all_targets)
```

```python
# --- Confusion matrix (counts) + metrics without extra imports ---
num_classes = len(full_train.classes)
cm = np.zeros((num_classes, num_classes), dtype=int)
for t, p in zip(y_true, y_pred):
    cm[t, p] += 1

acc = cm.trace() / cm.sum()
print(f"Test Accuracy: {acc:.3f}")
```

```python
# macro-precision/recall/F1 from CM
per_class_f1 = []
for i in range(num_classes):
    tp = cm[i, i]
    fp = cm[:, i].sum() - tp
    fn = cm[i, :].sum() - tp
    prec = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    rec  = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    f1   = (2 * prec * rec) / (prec + rec) if (prec + rec) > 0 else 0.0
    per_class_f1.append(f1)

macro_f1 = float(np.mean(per_class_f1))

print(f"Test Macro-F1: {macro_f1:.3f}")
```

```python
# Confusion Matrix (plot, counts)
plt.figure(figsize=(6,6))
im = plt.imshow(cm, cmap="Blues")
plt.title("Confusion Matrix (counts)")
plt.colorbar(im, fraction=0.046, pad=0.04)

ticks = np.arange(num_classes)
plt.xticks(ticks, full_train.classes, rotation=45, ha="right")
plt.yticks(ticks, full_train.classes)

thr = cm.max() * 0.6
for i in range(num_classes):
    for j in range(num_classes):
        plt.text(j, i, cm[i, j],
                 ha="center", va="center",
                 color=("white" if cm[i, j] > thr else "black"), fontsize=9)

plt.xlabel("Predicted"); plt.ylabel("True")
plt.tight_layout(); plt.show()
```

```python
model.eval()

misclassified = []
with torch.no_grad():
    for imgs, labels in test_dl:
        imgs   = imgs.to(device)     # <-- –ø–µ—Ä–µ–Ω–µ—Å–µ–Ω–Ω—è –Ω–∞ —Ç–æ–π —Å–∞–º–∏–π device, —â–æ –π –º–æ–¥–µ–ª—å
        labels = labels.to(device)

        logits = model(imgs)
        preds  = torch.argmax(logits, 1)

        # –∑–±–∏—Ä–∞—î–º–æ –ø—Ä–∏–∫–ª–∞–¥–∏, –¥–µ —î –ø–æ–º–∏–ª–∫–∞
        wrong_mask = preds != labels
        if wrong_mask.any():
            for img, y_true, y_pred in zip(imgs[wrong_mask], labels[wrong_mask], preds[wrong_mask]):
                misclassified.append((img.cpu(), int(y_true.cpu()), int(y_pred.cpu())))


# –≤—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è –ø–µ—Ä—à–∏—Ö 9 –ø–æ–º–∏–ª–∫–æ–≤–∏—Ö –≤–∏–ø–∞–¥–∫—ñ–≤
classes = train_ds.dataset.classes

# –ú–∞–ª—é—î–º–æ –ø–µ—Ä—à—ñ 9 –ø–æ–º–∏–ª–∫–æ–≤–∏—Ö –≤–∏–ø–∞–¥–∫—ñ–≤
n = min(9, len(misclassified))
fig, axes = plt.subplots(3, 3, figsize=(8, 8))
for i, ax in enumerate(axes.flat[:n]):
    img, true_label, pred_label = misclassified[i]
    ax.imshow(img.permute(1, 2, 0))
    ax.set_title(f"True: {classes[true_label]}\nPred: {classes[pred_label]}")
    ax.axis("off")
plt.tight_layout()
plt.show()
```
